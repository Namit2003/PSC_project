```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt

df = pd.read_csv('us1.csv')
# df = df.drop(columns=['AIRLINE_ID','CARRIER_NAME','ORIGIN_CITY_NAME','ORIGIN','DEST','ORIGIN_STATE_NM','ORIGIN_COUNTRY_NAME','DEST_STATE_ABR','DEST_STATE_NM','DEST_COUNTRY_NAME'])
# df.replace('', np.nan, inplace=True)
# df.dropna(inplace=True)
# df['MONTH'] = df['MONTH'].astype(int)

dic = {}
dic2 = {}

with open('us1.csv', 'r') as csvfile:
    csvreader = csv.reader(csvfile)
    count = 0
    for row in csvreader:
        #         removing 1st line
        if not count:
            count += 1
            continue
        if int(row[5]) in dic:
            dic[int(row[5])] += int(row[0])
        else:
            dic[int(row[5])] = int(row[0])

dic = dict(sorted(dic.items()))
# print(dic)
plt.plot(list(dic.keys()), list(dic.values()), color='orange')
plt.xticks(list(dic.keys()), rotation=90)
plt.xlabel("Years", fontweight='bold')
plt.ylabel("No. of passengers in x10 crores", fontweight='bold')
plt.show()

# df.to_csv('us1.csv', index=False)
# df.head()
```

```python
list_of_passengers = list(dic.values())
years = list(dic.keys())[1:]
rate_of_growth = []

for i in range(len(list_of_passengers)-1):
    rate_of_growth.append(
        (list_of_passengers[i+1]-list_of_passengers[i])*100/list_of_passengers[i])

plt.plot(years, rate_of_growth)
plt.xticks(years, rotation=90)
plt.xlabel("Years", fontweight='bold')
plt.ylabel("Rate of growth in passengers (%)", fontweight='bold')
plt.show()
```
